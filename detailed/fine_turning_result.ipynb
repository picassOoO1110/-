```python
import pandas as pd
import json

file_path = '/root/autodl-tmp/sft/gemma/lora/sft/predict_seniority/generated_predictions.jsonl'

with open(file_path, 'r', encoding='utf-8') as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data)

df['label_clean'] = df['label'].str.strip()

df['is_correct'] = df['predict'] == df['label_clean']

correct_count = df['is_correct'].sum()
incorrect_count = (~df['is_correct']).sum()
total_count = len(df)

accuracy = correct_count / total_count * 100
print(f"total number: {total_count}")
print(f"correct: {correct_count}")
print(f"wrong: {incorrect_count}")
print(f"accuracy: {accuracy:.2f}%")


incorrect_samples = df[~df['is_correct']][['predict', 'label_clean']].head(10)
print("\nsome wrong case：")
print(incorrect_samples)

# 7. 可选保存错误样本到 CSV
# incorrect_samples.to_csv('/mnt/data/incorrect_samples.csv', index=False)
```
